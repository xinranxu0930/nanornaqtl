```python
import pysam

bam_file = "/mnt/hpc/home/xuxinran/DirectSeq/data/zhaolin_240206/240201-zhaolin-RNA-merge/v0.9.6/map_bam_fastq/old/old_1_3mods_calls_sorted_map16.bam"
chrom = "chr1"
pos = 1533796

with pysam.AlignmentFile(bam_file, "rb") as bam:
    print(f"=== 测试位点: {chrom}:{pos} ===\n")
    
    # 方法1: fetch
    print("【方法1: fetch】")
    for read in bam.fetch(chrom, pos, pos+1):
        print(f"  read: {read.query_name}")
        print(f"    mapq: {read.mapping_quality}")
        print(f"    start: {read.reference_start}, end: {read.reference_end}")
    
    # 方法2: pileup
    print("\n【方法2: pileup (min_mapq=0)】")
    for pileup_column in bam.pileup(chrom, pos, pos+1, min_mapping_quality=0, truncate=True):
        if pileup_column.pos == pos:
            print(f"  pileup_column.n = {pileup_column.n}")
            for read in pileup_column.pileups:
                print(f"  read: {read.alignment.query_name}, mapq: {read.alignment.mapping_quality}, is_del: {read.is_del}, is_refskip: {read.is_refskip}")
```
